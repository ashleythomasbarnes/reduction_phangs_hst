```python
from astropy.table import Table, join, vstack
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
from astropy import constants as c
from astropy import units as u
import numpy as np

plt.style.use('paper')
```

```python
galaxies = ['ic5332',
            'ngc0628',
            'ngc5068',
            'ngc4535',
            'ngc4321',
            'ngc4303',
            'ngc4254',
            'ngc3627',
            'ngc3351',
            'ngc2835',
            'ngc1672',
            'ngc1566',
            'ngc1512',
            'ngc1433',
            'ngc1385',
            'ngc1365',
            'ngc1300',
            'ngc1087',
            'ngc7496']
```

```python
t_f657n = Table.read('./data/HST_WFC3_UVIS1.F657N.dat', format='ascii')
t_f658n = Table.read('./data/HST_WFC3_UVIS1.F658N.dat', format='ascii')
t_f658n_acs = Table.read('./data/HST_ACS_WFC.F658N.dat', format='ascii')
t_sample = Table.read('./data/phangs_sample_table_v1p6.fits')
t_sample = vstack([t_sample[t_sample['name'] == name] for name in galaxies])
t_sample.sort('name')
```

```python
t_sample['orient_vlsr_Ha_reshift'] = (((t_sample['orient_vlsr'].quantity/c.c.to('km/s')) * 6564.614) + 6564.614) * u.AA
t_sample.sort('orient_vlsr_Ha_reshift')
t_sample['name', 'orient_vlsr','orient_vlsr_Ha_reshift']
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)

ax.plot(t_f658n_acs['col1'], t_f658n_acs['col2'], zorder=1, alpha=0.5)
ax.fill_between(t_f658n_acs['col1'], t_f658n_acs['col2'], alpha=0.5, zorder=1)

ax.plot(t_f658n['col1'], t_f658n['col2'], zorder=2, alpha=0.5)
ax.fill_between(t_f658n['col1'], t_f658n['col2'], alpha=0.5, zorder=2)

ax.set_xlim([6450, 6700])
ax.set_ylim([0, 0.45])

ax.set_xlabel('Wavelength [Angstrom]')
ax.set_ylabel('Transmission')

lines = ax.get_lines()
legend1 = ax.legend([lines[i] for i in [0,1]], ["F658N ACS", "F658N UVIS"], fontsize=8, loc='upper right', title='Filter')
ax.add_artist(legend1)

ax.grid(':', alpha=0.2, which='both')

ax.xaxis.set_minor_locator(MultipleLocator(2))
ax.xaxis.set_major_locator(MultipleLocator(20))

fig.savefig('./filters_uvis_acs.png', dpi=300, bbox_inches='tight', transparent=False)
```
