```python
import astropy.units as u
import numpy as np
import warnings 
import sys
import os
from glob import glob
```

```python
input_dir = '/Users/abarnes/Dropbox/work/Smallprojects/galaxies/data_hstha/'
output_dir = '/Users/abarnes/Dropbox/work/Smallprojects/galaxies/scratch/transfer/'
```

```python
input_files_f657n = glob(f'{input_dir}*/hst/*657*_exp_drc_sci.fits', recursive=True)
input_files_f657n_err = glob(f'{input_dir}*/hst/*657*_err_drc_wht.fits', recursive=True)

input_files_f658n = glob(f'{input_dir}*/hst/*658*_exp_drc_sci.fits', recursive=True)
input_files_f658n_err = glob(f'{input_dir}*/hst/*658*_err_drc_wht.fits', recursive=True)

input_files_ha = glob(f'{input_dir}*/hst_contsub/*hst*ha_as2.fits', recursive=True)
input_files_ha_err = glob(f'{input_dir}*/hst_contsub/*hst*ha_err_as2.fits', recursive=True)

input_files_f65Xn = np.array(input_files_f657n + input_files_f658n).flatten()
input_files_f65Xn_err = np.array(input_files_f657n_err + input_files_f658n_err).flatten()
input_files_ha = np.array(input_files_ha)
input_files_ha_err = np.array(input_files_ha_err)
```

```python
gals = np.unique([f.removeprefix(input_dir).split('/')[0] for f in input_files_ha])
gals = np.sort(gals)
len(gals), gals
```

```python
gals_mosaics = ['ngc1300', 'ngc1672', 'ngc4254', 'ngc5068', 'ngc2835']
```

```python
for i in range(len(gals)):
# for i in [len(gals)-2]:

    gal = gals[i]
    print(f'Processing {gal}')
    
    if gal == 'ngc628mosaic':
        is_ngc628mosaic = True
        input_file_f65Xn = ''
        input_file_f65Xn_err = ''

    else: 
        is_ngc628mosaic = False
        input_file_f65Xn = [file for file in input_files_f65Xn if gal in file][0]
        input_file_f65Xn_err = [file for file in input_files_f65Xn_err if gal in file][0]

    input_file_ha = [file for file in input_files_ha if gal in file][0]
    input_file_ha_err = [file for file in input_files_ha_err if gal in file][0]

    if gal in gals_mosaics: 
        gal = f'{gal}mosaic'

    input_file_f65Xn_split = input_file_f65Xn.removeprefix(input_dir).split('/')[-1]

    if 'uvis' in input_file_f65Xn_split: 
        instrument = 'wfc3-uvis'
    elif 'acs' in input_file_f65Xn_split:
        instrument = 'acs-wfc'

    if 'f657' in input_file_f65Xn_split:
        filter = 'f657'
    elif 'f658' in input_file_f65Xn_split:
        filter = 'f658'

    if is_ngc628mosaic: 
        instrument = 'wfc3-uvis-acs-wfc'
        filter = 'f658'

    output_file_f65Xn = f'hlsp_phangs-hst_hst_{instrument}_{gal}_{filter}_v1_exp-drc-sci.fits'
    output_file_f65Xn_err = f'hlsp_phangs-hst_hst_{instrument}_{gal}_{filter}_v1_err-drc-wht.fits'

    output_file_ha = f'hlsp_phangs-hst_hst_{instrument}_{gal}_f555w-{filter}-f814w_v1_ha-img.fits'
    output_file_ha_err = f'hlsp_phangs-hst_hst_{instrument}_{gal}_f555w-{filter}-f814w_v1_ha-err.fits'

    output_file_f65Xn = f'{output_dir}{output_file_f65Xn}'
    output_file_f65Xn_err = f'{output_dir}{output_file_f65Xn_err}'

    output_file_ha = f'{output_dir}{output_file_ha}'
    output_file_ha_err = f'{output_dir}{output_file_ha_err}'

    if not is_ngc628mosaic: 
        print(f'{input_file_f65Xn} -> {output_file_f65Xn}')
        os.system(f'cp {input_file_f65Xn} {output_file_f65Xn}')
        print(f'{input_file_f65Xn_err} -> {output_file_f65Xn_err}')
        os.system(f'cp {input_file_f65Xn_err} {output_file_f65Xn_err}')

    print(f'{input_file_ha} -> {output_file_ha}')
    os.system(f'cp {input_file_ha} {output_file_ha}')
    print(f'{input_file_ha_err} -> {output_file_ha_err}')
    os.system(f'cp {input_file_ha_err} {output_file_ha_err}')
```
